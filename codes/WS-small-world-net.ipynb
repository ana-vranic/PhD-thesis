```python
import networkx as nx
import matplotlib.pyplot as plt
options = {
    'node_color': 'black',
    'node_size': 20,
    'width': 1,
}
fonts = 20
```

```python
fig, ax = plt.subplots(1, 3, )
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.1,
                    hspace=0.1)
plt.rcParams["figure.figsize"] = (20,6)


G = nx.watts_strogatz_graph(20,10,0.0)
ax[0].set_title(r'$p=0.0$', fontsize=fonts)

nx.draw_circular(G, **options, ax=ax[0])


G = nx.watts_strogatz_graph(20,10,0.05)
nx.draw_circular(G, **options, ax=ax[1])
ax[1].set_title(r'$p=0.05$', fontsize=fonts)


G = nx.watts_strogatz_graph(20,10,0.8)
nx.draw_circular(G, **options, ax=ax[2])
ax[2].set_title(r'$p=0.8$', fontsize=fonts)



plt.savefig("WS_model.pdf", bbox_inches="tight", dpi=200)
```
