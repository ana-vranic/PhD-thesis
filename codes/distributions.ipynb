```python
from scipy.stats import powerlaw
import numpy as np

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1)

a = 2.5

mean, var, skew, kurt = powerlaw.stats(a, moments='mvsk')

x = np.linspace(powerlaw.ppf(0.01, a),

                powerlaw.ppf(0.99, a), 100)

ax.plot(x, powerlaw.pdf(x, a),

       'r-', lw=5, alpha=0.6, label='powerlaw pdf')
```

```python
import powerlaw

theoretical_distribution = powerlaw.Lognormal(xmin=15, parameters=[0.3, 2.])
lognormal = theoretical_distribution.generate_random(100000)
powerlaw.plot_pdf(lognormal, linewidth=2, linestyle=':', label='lognormal')


theoretical_distribution = powerlaw.Power_Law(xmin=15, parameters=[2.5])
pl = theoretical_distribution.generate_random(100000)
powerlaw.plot_pdf(pl, linewidth=2, linestyle='-.', label='power-law')


theoretical_distribution = powerlaw.Exponential(xmin=15, parameters=[0.125])
exponential = theoretical_distribution.generate_random(100000)
powerlaw.plot_pdf(exponential, linewidth=2, linestyle='-.', label='exponential')

#plt.ylim(0.00001, 0.1)
#plt.xlim(xmax=1000)

plt.legend()
```

```python
powerlaw.plot_pdf(lognormal, linewidth=2, linestyle=':', label='lognormal')
powerlaw.plot_pdf(pl, linewidth=2, linestyle='-.', label='power-law')
plt.ylim(0.00001, 0.1)
```

```python
fit = powerlaw.Fit(lognormal, xmin=15)
R, p = fit.distribution_compare('power_law', 'lognormal', normalized_ratio=True)
print(R, p)
print('mu=', fit.lognormal.mu)
print('sigma=', fit.lognormal.sigma)
```

```python
fit = powerlaw.Fit(pl, xmin=15)
R, p = fit.distribution_compare('power_law', 'lognormal', normalized_ratio=True)
print(R, p)
print('gamma=', fit.power_law.alpha)
```

```python
fit = powerlaw.Fit(exponential, xmin=15)
R, p = fit.distribution_compare('exponential', 'lognormal', normalized_ratio=True)
print(R, p)
#print('gamma=', fit.power_law.alpha)
```
