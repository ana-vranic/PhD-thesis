```python
import powerlaw
import numpy as np
import matplotlib.pyplot as plt
import functions
```

```python
theoretical_distribution = powerlaw.Power_Law( parameters=[3])
pl = theoretical_distribution.generate_random(100000)
```

```python
theoretical_distribution = powerlaw.Exponential( parameters=[0.1])
exp = theoretical_distribution.generate_random(100000)
```

```python
theoretical_distribution = powerlaw.Lognormal( parameters=[3,0.5])
lognormal = theoretical_distribution.generate_random(100000)
```

```python
theoretical_distribution = powerlaw.Stretched_Exponential(parameters=[0.2, 2])
stexp = theoretical_distribution.generate_random(100000)
```

```python
theoretical_distribution = powerlaw.Truncated_Power_Law(parameters=[2.5, 0.1])
trpl = theoretical_distribution.generate_random(100000)
```

```python
import matplotlib as mpl

mpl.rcParams['axes.spines.right'] = False
mpl.rcParams['axes.spines.top'] = False


def plot_linear(data, title, params= " "):
    powerlaw.plot_pdf(data, linewidth=3, linestyle=':', color='black', label=params)
    plt.xscale("linear")
    plt.yscale("linear")
    plt.xlabel(r"$k$", fontsize=15)
    plt.ylabel(r"$P(k)$", fontsize=15)
    plt.title(title, fontsize=15)
    plt.legend(fontsize=14)

def plot_loglog(data, title):
    d = functions.bin_distribution(1, data)
    x, y = zip(*d)
    plt.loglog(x, y, 'o')
    powerlaw.plot_pdf(data, linewidth=2, linestyle='-', color='black', label="log-binned")
    plt.xlabel(r"$k$", fontsize=15)
    plt.ylabel(r"$P(k)$", fontsize=15)
    plt.legend(loc="lower left", fontsize=14)
    plt.title(title, fontsize=15)
```

```python
plt.figure(figsize = (10, 20))
plt.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=0.4, hspace=0.4)
i=5
j=2
c=1

plt.subplot(i, j, c)
plot_linear(exp, 'Exponential', r"$\lambda=0.1$")
c+=1

plt.subplot(i, j, c)
plot_loglog(exp, 'Exponential')
c+=1


plt.subplot(i, j, c)
plot_linear(pl, 'Power Law', r"$\gamma=3$")
c+=1

plt.subplot(i, j, c)
plot_loglog(pl, 'Power Law')
c+=1


plt.subplot(i, j, c)
plot_linear(lognormal, 'Log-normal', r"$\mu=3, \sigma=0.5 $")
c+=1

plt.subplot(i, j, c)
plot_loglog(lognormal, 'Log-normal')
c+=1

plt.subplot(i, j, c)
plot_linear(stexp, 'Stretched Exponential', r"$\lambda=0.2, \beta=2$")
c+=1

plt.subplot(i, j, c)
plot_loglog(stexp, 'Stretched Exponential')
c+=1

plt.subplot(i, j, c)
plot_linear(trpl, 'Power Law with Exponential Cutoff', r"$ \alpha=2.5, \lambda=0.2$")
c+=1

plt.subplot(i, j, c)
plot_loglog(trpl, 'Power Law with Exponential Cutoff')
c+=1

plt.savefig("Distributions_plots.pdf", bbox_inches="tight", dpi=200)
```
