```python
from scipy.stats import poisson
def poisson_funct(lambd):

    k_axis = np.arange(0, 80)
    distribution = np.zeros(k_axis.shape[0])
    for i in range(k_axis.shape[0]):
        distribution[i] = poisson.pmf(i, lambd)
        
    return k_axis, distribution
k_axis, distribution = poisson_funct(35)
plt.plot(k_axis,  distribution, '--', lw=2, label="Poisson($<k> = 50$)")
plt.xlabel("$k$", fontsize=font)
plt.ylabel("$P(k)$", fontsize=font)
plt.savefig("dist3.pdf", bbox_inches="tight", dpi=200)
```
